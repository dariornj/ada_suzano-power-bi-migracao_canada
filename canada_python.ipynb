```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importando os dados do dataset
# https://www.kaggle.com/datasets/ammaraahmad/immigration-to-canada

df_canada = pd.read_csv('canadian_immegration_data.csv')
df_canada
```

```python
# Importando os dados do dataset
# https://www.kaggle.com/datasets/tanuprabhu/population-by-country-2020

pop_country = pd.read_csv('population_by_country_2020.csv')
pop_country = pop_country.rename({'Country (or dependency)': 'Country',
                                 'Population (2020)': 'Population',}, axis='columns')
pop_country = pop_country.drop(['Yearly Change', 'Net Change', 'Density (P/Km²)', 'Fert. Rate', 'World Share'], axis = 1)
pop_country.head()                        
```

```python
#Checando que paises não possuem informação de população ou possuem nome diferente nas tabelas

checar_paises =  pop_country.merge(df_canada, on = 'Country', how = 'right', indicator = True)
checar_paises[checar_paises['_merge'] == 'right_only']['Country']
```

```python
# Corrigindo os nomes divergentes das bases de dados de população e imigração pro canadá

id = [21, 25, 37, 38, 47, 48, 49, 81, 94, 140, 141, 143, 145, 147, 150, 166, 169, 172, 183, 184, 185, 189, 190]
nomes = ['Bolivia', 'Brunei', 'China', 'China', 'Czech Republic (Czechia)', 'North Korea', 'DR Congo', 'Iran', 
        'Laos', 'South Korea', 'Moldova', 'Russia', 'Saint Kitts & Nevis', 'St. Vincent & Grenadines', 'Sao Tome & Principe',
        'Eswatini', 'Syria', 'North Macedonia', 'United Kingdom', 'Tanzania', 'United States', 'Venezuela', 'Vietnam']

i = 0
for number in id:
       df_canada.at[number, 'Country'] = nomes[i]
       i += 1
```

```python
# Separando os dados da database em informações e valores

df_canada_values = df_canada.drop(['Continent', 'Region', 'DevName'], axis = 1)
pop_country = pop_country.sort_values(by='Country', ignore_index=True)
df_canada_values['Population'] = pop_country['Population']
df_canada_values = df_canada_values.melt(['Country', 'Population'], var_name = 'Year', value_name='Immigrants')
df_canada_values.head() 
```

```python
anos = []
for year in range(1980, 2014):
    anos.append(str(year))

df_canada_info = df_canada.drop(anos, axis = 1)
df_canada_info = df_canada_info.drop('Total', axis = 1)
df_canada_info.head()
```

```python
df_canada_info = df_canada_info.merge(pop_country, on = 'Country', suffixes=None)
df_canada_info
```

```python
df_canada_values.to_csv(r'df_canada_values.csv', index=False, header=True)
df_canada_info.to_csv(r'df_canada_info.csv', index=False, header=True)
```
